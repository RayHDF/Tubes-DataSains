```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('RTA.csv')
```

```python
df
```

```python
# Replace Unknown with Male or Female
import random

def replace_unknown(value):
    if value == 'Unknown':
        return random.choice(['Male', 'Female'])
    else:
        return value

df['Sex_of_driver'] = df['Sex_of_driver'].apply(replace_unknown)
```

```python
# Binary Encoder
df['Sex_of_driver'] = df['Sex_of_driver'].replace({'Male': 0, 'Female': 1})
```

```python
# Value counts 'Age_band_of_driver'
df['Age_band_of_driver'].value_counts()
```

```python
# Replace 'Unknown' with random values

def replace_unknown(value):
    if value == 'Unknown':
        return random.choice(['18-30', '31-50', 'Over 51', 'Under 18'])
    else:
        return value

df['Age_band_of_driver'] = df['Age_band_of_driver'].apply(replace_unknown)
```

```python
# Value counts 'Age_band_of_driver'
df['Age_band_of_driver'].value_counts()
```

```python
# Label Encoder
df['Age_band_of_driver'] = df['Age_band_of_driver'].replace({'Under 18': 0, '18-30': 1, '31-50': 2, 'Over 51': 3})
```

```python
df['Age_band_of_driver'].value_counts()
```

```python
# Value counts Educational_level
df['Educational_level'].value_counts()
```

```python
def replace_unknown(value):
    if value == 'Unknown':
        return random.choice(['Junior high school', 'Elementary school', 'High school', 'Above high school', 'Writing & reading', 'Illiterate'])
    else:
        return value

df['Educational_level'] = df['Educational_level'].apply(replace_unknown)
```

```python
# Value Counts
df['Educational_level'].value_counts()
```

```python
# Label Encoder
df['Educational_level'] = df['Educational_level'].replace({'Illiterate': 0, 'Elementary school': 1, 'Junior high school': 2, 'High school': 3, 'Above high school': 4, 'Writing & reading': 5})
```

```python
# Value Counts
df['Educational_level'].value_counts()
```

```python
def replace_unknown(value):
    if value == 'Unknown' or value == 'unknown':
        return random.choice(['5-10yr', '2-5yr', 'Above 10yr', '1-2yr', 'Below 1yr', 'No Licence'])
    else:
        return value

df['Driving_experience'] = df['Driving_experience'].apply(replace_unknown)
```

```python
# Value Counts
df['Driving_experience'].value_counts()
```

```python
# Encoding
df['Driving_experience'] = df['Driving_experience'].replace({'No Licence': 0, 'Below 1yr': 1, '1-2yr': 2, '2-5yr': 3, '5-10yr': 4, 'Above 10yr': 5})
```

```python
# Replace Unknown with random values
def replace_unknown(value):
    if value == 'Unknown':
        return random.choice(['Two-way (divided with broken lines road marking)', 'Undivided Two way', 'Double carriageway (median)', 'One way', 'Two-way (divided with solid lines road marking)'])
    else:
        return value

df['Lanes_or_Medians'] = df['Lanes_or_Medians'].apply(replace_unknown)
```

```python
# Replace Unknown with random values
def replace_unknown(value):
    if value == 'Unknown':
        return random.choice(['Y Shape', 'No junction', 'Crossing', 'O Shape', 'T Shape', 'X Shape'])
    else:
        return value
    
df['Types_of_Junction'] = df['Types_of_Junction'].apply(replace_unknown)
```

```python
# Replace Unknown with random values
def replace_unknown(value):
    if value == 'Unknown':
        return random.choice(['Asphalt roads', 'Earth roads', 'Gravel roads', 'Asphalt roads with some distress'])
    else:
        return value
    
df['Road_surface_type'] = df['Road_surface_type'].apply(replace_unknown)
```

```python
# Replace Unknown with random values
def replace_unknown(value):
    if value == 'Unknown':
        return random.choice(['Vehicle with vehicle collision', 'Collision with roadside objects', 'Collision with pedestrians', 'Rollover', 'Collision with animals', 'Collision with roadside-parked vehicles', 'Fall from vehicles', 'With Train' ])
    else:
        return value
    
df['Type_of_collision'] = df['Type_of_collision'].apply(replace_unknown)

```

```python
# Value counts 'Vehicle_movement'
df['Vehicle_movement'].value_counts()
```

```python
# Replace Unknown with random values
def replace_unknown(value):
    if value == 'Unknown':
        return random.choice(['Going straight', 'Moving Backward', 'Reversing', 'Turnover', 'Getting off', 'Entering a junction', 'Overtaking', 'Waiting to go', 'U-Turn', 'Stopping', 'Parked'])
    else:
        return value
    
df['Vehicle_movement'] = df['Vehicle_movement'].apply(replace_unknown)
```

```python
df['Vehicle_movement'].value_counts()
```

```python
# Drop column 'Pedestrian_movement'
df = df.drop('Pedestrian_movement', axis=1)
```

```python
df['Cause_of_accident'].value_counts()
```

```python
# Replace Unknown with random values
def replace_unknown(value):
    if value == 'Unknown':
        return random.choice(['No distancing', 'Changing lane to the right', 'Changing lane to the left', 'Driving carelessly', 'No priority to vehicle', 'Moving Backward', 
                              'No priority to pedestrian', 'Overtaking', 'Driving under the influence of drugs', 'Driving to the left', 'Getting off the vehicle improperly',
                              'Driving at high speed', 'Overturning', 'Turnover', 'Overspeed', 'Overloading', 'Drunk driving' ,'Improper parking'])
    else:
        return value
    
df['Cause_of_accident'] = df['Cause_of_accident'].apply(replace_unknown)
```

```python
df
```

```python
# Apply One Hot Encoding to 'Lanes_or_Medians', 'Types_of_Junction', 'Road_surface_type', 'Type_of_collision', 'Vehicle_movement', 'Cause_of_accident'
df = pd.get_dummies(df, columns=['Lanes_or_Medians', 'Types_of_Junction', 'Road_surface_type', 'Type_of_collision', 'Vehicle_movement', 'Cause_of_accident'])
```
